```python
#Q20
df=pd.read_csv("D:\\ExcelR\\DS Assignments\\Assignment -1\\cars.csv")
df
import pandas as pd
from scipy.stats import norm

# Assuming `Cars` is a DataFrame with a column `MPG`
MPG = df['MPG']

# a. P(MPG>38)
P_MPG_gt_38 = 1 - norm.cdf(38, MPG.mean(), MPG.std())

# b. P(MPG<40)
P_MPG_lt_40 = norm.cdf(40, MPG.mean(), MPG.std())

# c. P(20<MPG<50)
P_MPG_20_to_50 = norm.cdf(50, MPG.mean(), MPG.std()) - norm.cdf(20, MPG.mean(), MPG.std())

print("a. P(MPG>38):", P_MPG_gt_38)
print("b. P(MPG<40):", P_MPG_lt_40)
print("c. P(20<MPG<50):", P_MPG_20_to_50)
''' 34.7% of cars in the dataset have a MPG greater than 38.
    72.9% of cars in the dataset have a MPG less than 40.
    89.8% of cars in the dataset have a MPG between 20 and 50.
'''
```

```python
#Q21.a
from scipy import stats
import matplotlib.pyplot as plt
MPG = df['MPG']

# Q-Q plot
stats.probplot(MPG, dist="norm", plot=plt)
plt.title('Q-Q plot for MPG')
plt.show()

# Normality test
stat, p = stats.normaltest(MPG)
print("Normality Test p-value:", p)
if p < 0.05:
    print("MPG does not follow a normal distribution")
else:
    print("MPG follows a normal distribution")
'''MPG of cars follows normal distribution'''
```

```python
#Q21.b
df=pd.read_csv("D:\\ExcelR\\DS Assignments\\Assignment -1\\wc-at.csv")
from scipy import stats
import matplotlib.pyplot as plt

# Assuming `wc_at` is a DataFrame with columns `AT` and `Waist`
AT = df['AT']
Waist = df['Waist']

# Q-Q plot for Adipose Tissue (AT)
stats.probplot(AT, dist="norm", plot=plt)
plt.title('Q-Q plot for Adipose Tissue (AT)')
plt.show()

# Normality test for Adipose Tissue (AT)
stat_AT, p_AT = stats.normaltest(AT)
print("Normality Test p-value for AT:", p_AT)
if p_AT < 0.05:
    print("AT does not follow a normal distribution")
else:
    print("AT follows a normal distribution")

# Q-Q plot for Waist Circumference (Waist)
stats.probplot(Waist, dist="norm", plot=plt)
plt.title('Q-Q plot for Waist Circumference (Waist)')
plt.show()

# Normality test for Waist Circumference (Waist)
stat_Waist, p_Waist = stats.normaltest(Waist)
print("Normality Test p-value for Waist:", p_Waist)
if p_Waist < 0.05:
    print("Waist does not follow a normal distribution")
else:
    print("Waist follows a normal distribution")
'''This data set is not normal because the data points follows an abnormalcurve'''
```
