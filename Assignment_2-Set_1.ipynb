```python
#Q1
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a DataFrame
df = pd.DataFrame({'Company Name': ['Allied Signal', 'Bankers Trust', 'General Mills', 'ITT Industries', 'J.P.Morgan & Co.', 'Lehman Brothers', 'Marriott', 'MCI', 'Merrill Lynch', 'Microsoft', 'Morgan Stanley', 'Sun Microsystems','Travelers','US Airways','Warner-Lambert'],
                   'Measure X': [24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39, 40.26, 32.95, 91.36, 25.99,39.42,26.71,35.00]})

# Plot the data
df.plot(kind='bar', x='Company Name', y='Measure X', figsize=(10, 6))
plt.xlabel('Company Name')
plt.ylabel('Measure X')
plt.title('Measure X by Company')
plt.show()

# Find the outliers
Q1 = df['Measure X'].quantile(0.25)
Q3 = df['Measure X'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[~((df['Measure X'] >= lower_bound) & (df['Measure X'] <= upper_bound))]

# Print the outliers
print('Outliers:')
print(outliers)

# Calculate the mean, standard deviation, and variance
mean = df['Measure X'].mean()
std = df['Measure X'].std()
variance = df['Measure X'].var()

# Print the results
print('Mean:', mean)
print('Standard deviation:', std)
print('Variance:', variance)
'''    Outliers:
      Company Name    Measure X
10  Morgan Stanley      91.36
Mean: 33.27133333333333
Standard deviation: 16.945400921222028
Variance: 287.1466123809524'''
```
